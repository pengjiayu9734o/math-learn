```python
#小结练习
draw3d(
    Points3D((-1,-2,2)),
    Arrow3D((-1,-2,2)),
    Box3D(-1,-2,2)
)

pml=[-1,1]
vertices=[(x,y,z) for x in pml for y in pml for z in pml]
edges=[((-1,y,z),(1,y,z)) for y in pml for z in pml]+\
        [((x,-1,z),(x,1,z)) for x in pml for z in pml]+\
        [((x,y,-1),(x,y,1)) for x in pml for y in pml]
draw3d(
    Points3D(*vertices),
    *[Segment3D(*edge) for edge in edges]
)
```

```python
def add(*vectors):
    #zip函数将列表中元组对应维度的值给放到新的元组里面，结果是对应维度的值组成的元组列表
    by_coordinate=zip(*vectors)
    coordinate_sums=[sum(coords) for coords in by_coordinate]
    return tuple(coordinate_sums)
```

```python
#3.2.6练习
a1=Arrow3D((4,0,3),color=blue)
a2=Arrow3D((-1,0,1),color=red)
v=add(*[(4,0,3),(-1,0,1)])
print(v)
a3=Arrow3D(tip=v,tail=(4,0,3),color=green)
draw3d(Arrow3D((4,0,3),color=blue),
       Arrow3D((-1,0,1),color=red),
       Arrow3D(v,color=purple),
       Arrow3D(tip=v,tail=(4,0,3),color=green))


```
